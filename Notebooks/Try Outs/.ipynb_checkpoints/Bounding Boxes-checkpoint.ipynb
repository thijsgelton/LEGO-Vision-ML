```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
bounding_boxes_file = r"C:\Users\Thijs\CNTK-Samples\Examples\Image\DataSets\Grocery\positive\WIN_20160803_11_29_07_Pro.bboxes.tsv"
corresponding_image_file = r"C:\Users\Thijs\CNTK-Samples\Examples\Image\DataSets\Grocery\positive\WIN_20160803_11_29_07_Pro.jpg"
```

```python
with open(bounding_boxes_file) as bounding_boxes:
    # Create figure and axes
    fig,ax = plt.subplots(1)

    # Display the image
    ax.imshow(plt.imread(corresponding_image_file))

    # Create a Rectangle patch
    for line in bounding_boxes.readlines():
        x, y, h, w = [int(number.strip('\n')) for number in line.split('\t')]
        rect = patches.Rectangle((x,y),w,h,linewidth=1,edgecolor='r',facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```
