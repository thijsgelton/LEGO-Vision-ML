```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
bounding_boxes_file = r"D:\LEGO Vision Datasets\detection-synthetic-data\positive\5.bboxes.tsv"
corresponding_image_file = r"D:\LEGO Vision Datasets\detection-synthetic-data\positive\5.jpg"
```

```python
with open(bounding_boxes_file) as bounding_boxes:
    # Create figure and axes
    fig,ax = plt.subplots(1)

    # Display the image
    ax.imshow(plt.imread(corresponding_image_file))

    # Create a Rectangle patch
    for line in bounding_boxes.readlines():
        x_min, y_min, x_max, y_max = [int(number.strip('\n')) for number in line.split('\t')]
        
        rect = patches.Rectangle((x_min, y_min), abs(x_max), abs(y_max),linewidth=1,edgecolor='r',facecolor='none')

        # Add the patch to the Axes
        ax.add_patch(rect)

    plt.show()
```
